```python
import pandas as pd
```

```python
# Cargamos el dataframe original 
df = pd.read_csv('Restaurant reviews.csv')
# Filtramos las filas del dataset tal que la columna 'Metadata' contiene el string la expresión 'Follower' y no contiene 
# la expresión 'Review'
df['Metadata'] = df['Metadata'].fillna('')
filtro = df['Metadata'].str.contains('Follower') & ~df['Metadata'].str.contains('Review')
resultados = df[filtro]['Metadata']
# Mostrar las filas que cumplen con la condición
print(resultados.head(30))
```

Vemos que no hay ninguna fila en la que aparezca únicamente el número de followers, todas las filas contienen el número de Reviews/Posts

```python
# Filtramos las filas del dataset tal que la columna 'Metadata' no contiene el string la expresión 'Follower' y sí
# contiene la expresión 'Review'
filtro = ~df['Metadata'].str.contains('Follower') & df['Metadata'].str.contains('Review')
resultados = df[filtro]['Metadata']
# Mostrar las filas que cumplen con la condición
print(resultados.head(30))
```

Observamos como hay filas en las que solamente aparece el número de reviews en la columna Metadata

Para obtener las columnas Reviews y Followers a partir de Metadata lo que haremos será seleccionar la primera sucesión de dígitos que aparezca en Metadata y asignarla a Reviews. En caso de que haya una segunda sucesión de dígitos, se asignará a Followers.

```python
# Extraemos el número de Reviews y Followers de la columna Metadata en dos nuevas columnas
df[['Reviews', 'Followers']] = df['Metadata'].str.extract('(\d+)\D*(\d*)', expand=True)

# Convertimos las nuevas columnas a tipos numéricos
df[['Reviews', 'Followers']] = df[['Reviews', 'Followers']].apply(pd.to_numeric)
```
